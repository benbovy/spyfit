```python
ds = spy.load_dataset('sfit4.ctl', fmt='sfit4')
```

```python
ds_ref_profiles = xr.Dataset({name: da for name, da in ds.items()
                              if name.startswith('reference__')})
```

```python
map_gas_id_name = {da.attrs.get('gas_index'): name.split('__')[-1]
                   for name, da in ds_ref_profiles.items()
                   if da.attrs.get('gas_index', False)}
for i in range(1, 100):
    if i not in map_gas_id_name.keys():
        map_gas_id_name[i] = 'OTHER'
        
#map_gas_id_name
```

```python
spy.io.sfit4.write_reference(ds_ref_profiles, 'reference_test.prf',
                             map_gas_id_name)
```
